### Create correlation matrices and heatmaps to identify relationships between variables.

```python
df1 = df.drop(columns =['area_type','availability','location', 'size','society','total_sqft'])
df1
```

```python
# hep = df.loc[df.price == 'hepatitis']
corr_mat_hep = df1.corr().round(2)
corr_mat_hep
```

```python
plt.figure(figsize=(6,6))
plot = sns.heatmap(df1.corr().round(2), annot = True)
```

### Outlier

```python
# Assuming 'df' is your DataFrame
sns.displot(df['bath'], kde=True)
plt.xlabel('Number of Bathrooms')
plt.ylabel('Density')
plt.title('Distribution of Number of Bathrooms')
plt.show()
```

```python
sns.boxplot(df['bath'])
```

```python
upper_limit = df['bath'].mean() + 3* df['bath'].std()
lower_limit = df['bath'].mean() - 3* df['bath'].std()
print("upper_limit" , upper_limit)
print("lower_limit" , lower_limit)
```

```python
df.loc[(df['bath'] > upper_limit) | (df['bath'] <lower_limit)]
```

```python
new_df = df.loc[(df['bath'] > upper_limit) | (df['bath'] <lower_limit)]
print("old data" , len(df))
print("new data", len(new_df))
```
